```python
import pandas as pd
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

folder_path = "/content/drive/MyDrive/DATN/data/daily"

# Lấy danh sách file CSV
csv_files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

for file in csv_files:
    file_path = os.path.join(folder_path, file)

    # Extract latitude and longitude from the first two lines of each CSV
    latitude = pd.NA
    longitude = pd.NA

    try:
        with open(file_path, 'r') as f:
            meta_header_line = f.readline().strip() # e.g., "latitude\tlongitude\t..."
            meta_values_line = f.readline().strip() # e.g., "10.720562\t105.10601\t..."

        # Determine delimiter (assume tab or comma)
        delimiter = '\t' if '\t' in meta_header_line else ','

        meta_headers = [h.strip().lower() for h in meta_header_line.split(delimiter)]
        meta_values = [v.strip() for v in meta_values_line.split(delimiter)]

        latitude_idx = meta_headers.index('latitude')
        longitude_idx = meta_headers.index('longitude')

        latitude = float(meta_values[latitude_idx])
        longitude = float(meta_values[longitude_idx])

    except (ValueError, IndexError, FileNotFoundError) as e:
        print(f"Error: Could not extract latitude/longitude from {file_path}. Setting to NA. Error: {e}")

    # 1. Đọc file, bỏ 3 dòng đầu (metadata header, metadata values, blank line)
    df = pd.read_csv(
        file_path,
        header=None,
        skiprows=3, # Now skips metadata header, values, and blank line, starting from 'time' header
        engine="python"
    )

    # 2. Lấy dòng đầu tiên sau khi skip làm header
    df.columns = df.iloc[0].astype(str).str.strip()

    # 3. Bỏ dòng header cũ
    df = df.iloc[1:].reset_index(drop=True)

    # Add latitude and longitude columns to the time-series data
    df['latitude'] = latitude
    df['longitude'] = longitude

    # 4. Lưu lại đúng tên file (ghi đè)
    df.to_csv(file_path, index=False)

    print(f"Đã xử lý xong: {file}")

print("Hoàn tất xử lý toàn bộ file trong folder")
```

```python
import pandas as pd
import os

FOLDER_PATH = "/content/drive/MyDrive/DATN/data/daily"   # Adjusted to point to the processed daily data
OUTPUT_PATH = "/content/drive/MyDrive/vietnam_weather_daily_latlong.csv"

all_dfs = []

for file in os.listdir(FOLDER_PATH):
    if not file.endswith(".csv"): # Ensure only CSVs are processed
        continue

    file_path = os.path.join(FOLDER_PATH, file)
    df = pd.read_csv(file_path)

    # Chuẩn hoá tên cột thời gian
    if "time" in df.columns:
        df["time"] = pd.to_datetime(df["time"])
        df = df.rename(columns={"time": "date"})

    # The 'province_id' mapping is no longer needed as latitude/longitude are in the files
    # We also do not add 'province_id' as a column here based on the user's latest request.

    all_dfs.append(df)

# Gộp tất cả tỉnh
merged_df = pd.concat(all_dfs, ignore_index=True)

# Sắp xếp cho đúng time series (removed province_id from sort_values)
merged_df = merged_df.sort_values(["date", "latitude", "longitude"]).reset_index(drop=True)

# Remove 'province_id' if it was added previously (this handles cases where it might still exist)
if 'province_id' in merged_df.columns:
    merged_df = merged_df.drop(columns=['province_id'])

# Lưu file
merged_df.to_csv(OUTPUT_PATH, index=False)

print("✅ Gộp dữ liệu 63 tỉnh thành công")
print("Shape:", merged_df.shape)
```
