```python
import pandas as pd
import os
from google.colab import drive

drive.mount('/content/drive')

folder_path = "/content/drive/My Drive/DATN/data/hourly"
csv_files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

dataframes = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)

    # Đọc bằng python engine, bỏ 3 dòng đầu
    df = pd.read_csv(
        file_path,
        header=None,
        skiprows=3,
        engine="python"
    )

    # Lấy dòng đầu làm header
    df.columns = df.iloc[0]

    # Bỏ dòng header cũ
    df = df.iloc[1:].reset_index(drop=True)

    # Làm sạch tên cột (rất nên)
    df.columns = df.columns.astype(str).str.strip()

    dataframes.append(df)

merged_df = pd.concat(dataframes, ignore_index=True)

output_path = os.path.join(folder_path, "63tinh_hourly.csv")
merged_df.to_csv(output_path, index=False)

print("Gộp file thành công")
```
